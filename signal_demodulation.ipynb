由上节知, 调幅波的特点是载波的振幅受调制信号的控制作周期性的变化。 这个变化的周期与调制信号的周期相同, 而振幅变化则与调制信号的振幅成正 比。现在进一步分析调幅波的特点。
为简化分析手续起见,假定调制信号是简谐振荡, 其表示式为
$$
v_{\Omega}=V_{\Omega} \cos \Omega t
$$
如果用它来对载波 $v=V_0 \cos \omega_0 t$ 进行调幅, 那么, 在理想的情况下, 已调波的振 幅为
$$
V(t)=V_0+k_{\mathrm{a}} V_{\Omega} \cos \Omega t
$$
式中, $k$ 为比例常数。
因此,已调波可以用下式表示:
$$
\begin{aligned}
v(t) & =V(t) \cos \omega_0 t \\
& =\left(V_0+k_{\mathrm{a}} V_{\Omega} \cos \Omega t\right) \cos \omega_0 t \\
& =V_0\left(1+m_{\mathrm{a}} \cos \Omega t\right) \cos \omega_0 t
\end{aligned}
$$
式中, $m_{\mathrm{a}}=\frac{k_{\mathrm{a}} V_{\Omega}}{V_0}$ 叫做调幅指数 ( amplitude modulation factor) 或调幅度, 它通常以百分数来表示。

```python
point=8192
n=np.array(range(point))
v0=0.1
v_omega=0.01
# ka=8
# ma=ka*v_omega/v0
ma=0.5
ka=ma*v0/v_omega
print(f'ka={ka:.2f}')
fs=10e6
fa=10e3
fcarrier=2e6
V_omega=0.01
```

```python
T=fs/fa
print("T=",T)
omega0=2*np.pi/T
print("omega0=","{:.5f}".format(omega0))
```

```python
modulating_wave=v_omega*np.sin(omega0*n)
plt.plot(modulating_wave)
plt.title('modulating_wave')
plt.show()
# carrier=np.sin()
```

```python
Tc=fs/fcarrier
print("Tc=",Tc)
omegac=2*np.pi/Tc
print("omegac=","{:.5f}".format(omegac))
```

```python

carrier=v0*np.sin(omegac*n)
plt.plot(carrier)
plt.title('carrier')
plt.show()
```

```python
modulated_wave=(v0+ka*modulating_wave)*carrier
plt.plot(modulated_wave)
plt.show()
```
